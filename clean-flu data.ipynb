```python
import os
import pandas as pd
import numpy as np

flu_csv = "Deaths_data.csv"
week_month_csv = "Week_to_month.csv"

flu_df = pd.read_csv(flu_csv)
week_month_df = pd.read_csv(week_month_csv)
flu_df.head(-10)
```

```python
flu_column_subset = pd.DataFrame(flu_df, columns = ["State", "season", "MMWR Year/Week", "Deaths from influenza", "Deaths from pneumonia", "Deaths from pneumonia and influenza", "All Deaths"])


flu_column_subset.head()
flu_column_subset["season"].count()
```

```python
# flu2017 = flu_column_subset[flu_column_subset['MMWR Year/Week]
flu_column_subset['MMWR Year/Week'] = flu_column_subset['MMWR Year/Week'].astype(str)
flu_column_subset['Year'] = flu_column_subset['MMWR Year/Week'].str[:4]
flu_column_subset['Week'] = flu_column_subset['MMWR Year/Week'].str[4:]

flu_final = pd.DataFrame(flu_column_subset, columns = ["State", "MMWR Year/Week", "Deaths from influenza", "Deaths from pneumonia", "Deaths from pneumonia and influenza", "All Deaths", "Year", "Week"])

flu_final.head()

flu_column_subset['MMWR Year/Week'] = flu_column_subset['MMWR Year/Week'].astype(str)
flu_final
```

```python
week_month_df['Year_week'] = week_month_df['Year_week'].astype(str)


week_month_df.dtypes
```

```python
merged_flu_week = pd.merge(flu_final, week_month_df, how = "left", left_on = "MMWR Year/Week", right_on = "Year_week")

merged_flu_week.head()
```

```python
grouped_df = merged_flu_week.groupby(["State", "Month"])

all_death = grouped_df["All Deaths"].mean()
sick_death = grouped_df["Deaths from pneumonia and influenza"].mean()
final_flu_data = pd.DataFrame({"sick death": sick_death, "all death": all_death})
final_flu_data
```

```python
final_flu_data.to_csv("Resources/flu_data.csv", index=True, header=True)
```
