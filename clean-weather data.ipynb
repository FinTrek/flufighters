```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
weather_data_to_load = "monthly_mean_temps.csv"
state_data_to_load = "State_Names.csv"

# Read the weahter and state Data
weather_df = pd.read_csv(weather_data_to_load, low_memory=False)
state_df = pd.read_csv(state_data_to_load, low_memory=False)
```

```python
weather_df.head()
```

```python
state_df.head()
```

```python
# Combine the data into a single dataset to get the full name of each state
merge_table = pd.merge(weather_df, state_df, left_on="State",right_on="State Code", how="left")
merge_table
```

```python
# group by state
grouped_df =merge_table.groupby(['US State Name','State'])
grouped_df
# set a blank list to contain monthly data later
frames=[]

for month in ["January","February","March","April","May","June","July","August","September","October","November","December"]:
    temp=grouped_df[month].mean()
    monthly_data = pd.DataFrame({
    "Month": month,
    "Avg_temp": temp})
    frames.append(monthly_data)

# consolidate all monthly data in frames list 
main_table = pd.concat(frames)
main_table
```

```python
main_table.to_csv("Resources/weather_data.csv", index=True, header=True)
```
